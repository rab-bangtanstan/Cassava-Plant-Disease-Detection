```python
import sys
import os
import pandas as pd
import glob
import shutil
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
df = pd.read_csv('H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\train.csv')
print(df)
```

```python
images_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\train_images"
```

```python
cbb_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\cassava classes\\Cassava Bacterial Blight"
cbsd_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\cassava classes\\Cassava Brown Streak Disease"
cgm_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\cassava classes\\Cassava Green Mottle"
cmd_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\cassava classes\\Cassava Mosaic Disease"
healthy_path = "H:\\PROJECT WORK -II\\datasets\\Cassava Ongoing\\cassava classes\\Healthy"
```

```python
cnt0=0
cnt1=0
cnt2=0
cnt3=0
cnt4=0
for ind in df.index:
    if df['label'][ind]==0:
        #print("cbb:"+str(df['label'][ind]))
        cnt0=cnt0+1
        
    elif df['label'][ind]==1:
        #print("cbsd: "+str(df['label'][ind]))
        cnt1=cnt1+1
        
    elif df['label'][ind]==2:
        #print("cgm: "+str(df['label'][ind]))
        cnt2=cnt2+1
        
    elif df['label'][ind]==3:
        #print("cmd: "+str(df['label'][ind]))
        cnt3=cnt3+1
        
    elif df['label'][ind]==4:
        #print("healthy:"+str(df['label'][ind]))
        cnt4=cnt4+1
        

        
print("Cassava Bacterial Blight images:" + str(cnt0))
print("Cassava Brwon Streak Disease images:" + str(cnt1))
print("Cassava Green Mottle images:" + str(cnt2))
print("Cassava Mossaic Disease images:" + str(cnt3))
print("Cassava Healthy images:" + str(cnt4))

print("Total images:" + str(cnt0+cnt1+cnt2+cnt3+cnt4))
        
    
```

```python
count =0
for ind in df.index:
    image_name = df['image_id'][ind]
    append_path = images_path + "\\" + image_name
    count=count+1
    if df['label'][ind]==0:
        src_path = append_path
        dest_path = cbb_path
        print("Copying files......."+str(count))
        shutil.copy(src_path,dest_path)
        
    elif df['label'][ind]==1:
        src_path = append_path
        dest_path = cbsd_path
        print("Copying files......."+str(count))
        shutil.copy(src_path,dest_path)
        
    elif df['label'][ind]==2:
        src_path = append_path
        dest_path = cgm_path
        print("Copying files......."+str(count))
        shutil.copy(src_path,dest_path)
        
    elif df['label'][ind]==3:
        src_path = append_path
        dest_path = cmd_path
        print("Copying files......."+str(count))
        shutil.copy(src_path,dest_path)
        
    elif df['label'][ind]==4:
        src_path = append_path
        dest_path = healthy_path
        print("Copying files......."+str(count))
        shutil.copy(src_path,dest_path)
    
        
```
